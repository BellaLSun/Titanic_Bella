```python
import pandas as pd #数据分析
import numpy as np #科学计算
from pandas import Series,DataFrame

data_train = pd.read_csv("Train.csv")
data_train.columns
```

```python
data_train.info()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(18,6), dpi=1600) 
# the alpha attribute is used to make semitransparent circle markers.
fig.set(alpha=0.2)  # 设定图表颜色alpha参数
alpha=alpha_scatterplot = 0.2 
alpha_bar_chart = 0.55

plt.subplot2grid((2,3),(0,0))             # 在一张大图里分列几个小图
data_train.Survived.value_counts().plot(kind='bar',alpha=alpha_bar_chart)# plots a bar graph of those who surived vs those who did not. 
plt.title("Survival by Age,  (1 = Survived)") # puts a title on our graph
plt.ylabel(u"人数")  

plt.subplot2grid((2,3),(0,1))
data_train.Pclass.value_counts().plot(kind="bar",alpha=alpha_scatterplot)
plt.ylabel(u"人数")
plt.title(u"乘客等级分布")

plt.subplot2grid((2,3),(0,2))
plt.scatter(data_train.Survived, data_train.Age, alpha=alpha_bar_chart)
plt.ylabel(u"年龄")                         # sets the y axis lable
plt.grid(b=True, which='major', axis='y') # formats the grid line style of our graphs
plt.title(u"按年龄看获救分布 (1为获救)")


plt.subplot2grid((2,3),(1,0), colspan=2)
data_train.Age[data_train.Pclass == 1].plot(kind='kde')   # plots a kernel desnsity estimate of the subset of the 1st class passanges's age
data_train.Age[data_train.Pclass == 2].plot(kind='kde')
data_train.Age[data_train.Pclass == 3].plot(kind='kde')
plt.xlabel(u"年龄")# plots an axis lable
plt.ylabel(u"密度") 
plt.title(u"各等级的乘客年龄分布")
plt.legend((u'头等舱', u'2等舱',u'3等舱'),loc='best') # sets our legend for our graph.


plt.subplot2grid((2,3),(1,2))
data_train.Embarked.value_counts().plot(kind='bar')
plt.title(u"各登船口岸上船人数")
plt.ylabel(u"人数")  
plt.show()

```

```python
#看看各乘客等级的获救情况
fig = plt.figure()

Survived_0 = data_train.Pclass[data_train.Survived == 0].value_counts()
Survived_1 = data_train.Pclass[data_train.Survived == 1].value_counts()
df=pd.DataFrame({u'survived':Survived_1, u'unsurvived':Survived_0})
df.plot(kind='bar', stacked=True,alpha=0.5)
plt.title(u"survived in different classes")
plt.xlabel(u"class") 
plt.ylabel(u"survived people") 

plt.show()
```

```python
#看看各性别的获救情况
fig = plt.figure()
  # 设定图表颜色alpha参数

Survived_m = data_train.Survived[data_train.Sex == 'male'].value_counts()
Survived_f = data_train.Survived[data_train.Sex == 'female'].value_counts()
df=pd.DataFrame({u'male':Survived_m, u'female':Survived_f})
df.plot(kind='bar', stacked=True,alpha=0.5)
plt.title(u"survived in different gender")
plt.xlabel(u"sex") 
plt.ylabel(u"survived people number")
plt.show()
```

```python
f1=data_train.Survived[data_train.Sex == 'female'][data_train.Pclass != 3].value_counts().sort_index()
```

```python
f2=data_train.Survived[data_train.Sex == 'female'][data_train.Pclass == 3].value_counts().sort_index()
```

```python
m1=data_train.Survived[data_train.Sex == 'male'][data_train.Pclass != 3].value_counts().sort_index()
```

```python
m2=data_train.Survived[data_train.Sex == 'male'][data_train.Pclass == 3].value_counts()
```

```python
#然后我们再来看看各种舱级别情况下各性别的获救情况
fig=plt.figure(figsize=(18,6))
fig.set(alpha=0.65) # 设置图像透明度，无所谓
plt.title("Who Survived according to sex and Pclass")

#'best': 0, (only implemented for axis legends)
ax1=fig.add_subplot(141)
f1.plot(kind='bar', label="female highclass", color='#FA2479')
ax1.set_xticklabels(["died", "servived"], rotation=0)
ax1.legend(["female/high"], loc='best')

ax2=fig.add_subplot(142, sharey=ax1)
f2.plot(kind='bar', label='female, low class', color='pink')
ax2.set_xticklabels(["died", "servived"], rotation=0)
plt.legend(["female/low"], loc='best')

ax3=fig.add_subplot(143, sharey=ax1)
m1.plot(kind='bar', label='male, high class',color='lightblue')
ax3.set_xticklabels(["died", "servived"], rotation=0)
plt.legend(["male/high"], loc='best')

ax4=fig.add_subplot(144, sharey=ax1)
m2.plot(kind='bar', label='male low class', color='steelblue')
ax4.set_xticklabels(["died", "servived"], rotation=0)
plt.legend(["male/low"], loc='best')

plt.show()
```
